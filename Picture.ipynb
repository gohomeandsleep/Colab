```python
import cv2
from google.colab.patches import cv2_imshow
img_path = '/content/sample_data/house.jpg'
img = cv2.imread(img_path, cv2.IMREAD_COLOR)
cv2_imshow(img)
```

```python
img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
img_g = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_g)
```

```python
img_edges = cv2.Canny(img_g, 50, 150, apertureSize = 3)
cv2_imshow(img_edges)
```

```python
img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
img_blur = cv2.blur(img,(3,3))
cv2_imshow(img_blur)
```

```python
img_g = cv2.cvtColor(img_blur, cv2.COLOR_BGR2GRAY)
img_edge = cv2.Canny(img_g, 50, 150, apertureSize = 3)
cv2_imshow(img_edge)
```
